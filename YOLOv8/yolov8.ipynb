### hyperparameter 정의!
1. source: source directory for images or videos
2. conf: object confidence threshold for detection
3. iou: intersection over union threshold for NMS
4. device: str type -> 나는 0으로 설정
5. hide_conf: hide confidence scores

```python
# learning from scratch!
model = YOLO("./runs/detect/train7/weights/best.pt")
model.train(data = "./VOC2028/data.yaml", optimizer = "AdamW", epochs = 30, batch = 4,  device = '0', val = True, verbose = True, lr0 = 1e-3, lrf = 1e-5, cos_lr = True)
```
